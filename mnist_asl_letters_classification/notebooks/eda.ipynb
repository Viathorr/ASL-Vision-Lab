# ASL Letters MNIST Dataset Exploration

## Importing Dependencies

```python
import os
import warnings
import pandas as pd

warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt  # Plotting library for visualization

viz_utils_path = os.path.abspath(os.path.join(os.getcwd(), "..", "..", "utils", "visualization.py"))

if os.path.exists(viz_utils_path):
  import sys
  sys.path.append(os.path.dirname(viz_utils_path))
  import visualization as viz
else:
  print("Visualization utils not found.")
```

## Data Loading

```python
data_path = os.path.abspath(os.path.join(os.getcwd(), "..", "data", "sign_mnist_train.csv"))

if not os.path.exists(data_path):
  print("Data not found.")
```

```python
df = pd.read_csv(data_path)

df.head()
```

## Visualization

```python
label_counts = df["label"].value_counts(normalize=True).sort_index()

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', 
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf',
          '#aec7e8', '#ffbb78', '#98df8a', '#ff9896', '#c5b0d5',
          '#c49c94', '#f7b6d2', '#c7c7c7', '#dbdb8d', '#9edae5',
          '#393b79', '#637939', '#8c6d31', '#843c39']

plt.figure(figsize=(8, 8))  

label_counts.plot.pie(autopct='%1.1f%%', startangle=90, labels=None, colors=colors) 
plt.title('Distribution of Labels')  
plt.ylabel('') 
plt.legend(labels=label_counts.index, title="Labels", bbox_to_anchor=(1, 1))

plt.show()
```
