# ASL Letters MNIST Dataset Exploration

## Importing Dependencies

```python
import os
import warnings
import pandas as pd

warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt  # Plotting library for visualization

try:
  import utils.visualization as viz
except:
  print("Visualization utils not found.")
```

## Data Loading

```python
train_data_path = os.path.abspath(os.path.join(os.getcwd(), "..", "data", "sign_mnist_train.csv"))
test_data_path = os.path.abspath(os.path.join(os.getcwd(), "..", "data", "sign_mnist_test.csv"))

if not os.path.exists(train_data_path) or not os.path.exists(test_data_path):
  print("Data not found.")
```

```python
df = pd.read_csv(train_data_path)
test_df = pd.read_csv(test_data_path)

df.head()
```

## Visualization

```python
train_label_counts = df["label"].value_counts(normalize=True).sort_index()
test_label_counts = test_df["label"].value_counts(normalize=True).sort_index()

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', 
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf',
          '#aec7e8', '#ffbb78', '#98df8a', '#ff9896', '#c5b0d5',
          '#c49c94', '#f7b6d2', '#c7c7c7', '#dbdb8d', '#9edae5',
          '#393b79', '#637939', '#8c6d31', '#843c39']

train_label_counts.plot.pie(autopct='%1.1f%%', startangle=90, labels=None, colors=colors, ax=ax[0])
ax[0].set_title('Train Data')
ax[0].set_ylabel('')

test_label_counts.plot.pie(autopct='%1.1f%%', startangle=90, labels=None, colors=colors, ax=ax[1])
ax[1].set_title('Test Data')
ax[1].set_ylabel('')

plt.legend(labels=train_label_counts.index, title="Labels", bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()

viz.save_figure(fig, "label_distribution.png")
```
